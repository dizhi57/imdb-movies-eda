## 5. Обработка колонок runtime и gross_earn: приведение к числовому типу

Приведем значения признаков runtime и gross_earn к числовому типу, исключив лишние символы

### *5.1. runtime*

Выведем Series ['runtime'] для детального просмотра данных

```python
dataset['runtime']
```

По индексу 4997 в признак попало аномальное значение - Drama, Romance

Выведем уникальные значения по Series ['runtime'] с помощью метода unique()

```python
df['runtime'].unique()
```

Таких аномальных значений несколько. Следует исключить их из датасета

```python
import re
```

Напишем функцию, которая исключит нецифровые символы с помощью регулярных выражений

```python
import re
import pandas as pd

def clear_runtime(runtime):
    """
    Удаляет все НЕцифровые символы (regex) и возвращает целое число.
    Если после очистки цифр нет — возвращает pd.NA.
    """
    if pd.isna(runtime):
        return pd.NA

    s = str(runtime)
    digits = re.sub(r"\D+", "", s)   # \D = любой НЕ-цифровой символ

    return int(digits) if digits else pd.NA
```

Примените функцию к Series ['runtime'] с помощью метода apply()

```python
df["runtime_clean"] = df["runtime"].apply(clear_runtime).astype("Int64")
```

Создадим столбец runtime_clear в исходном датасете, применив функцию к признаку runtime

```python
dataset['runtime_clear'] = dataset["runtime"].apply(clear_runtime).astype("Int64")
```

Выведем Series ['runtime_clear'] для оценки результата

```python
dataset['runtime_clear']
```

По индексу 4997, где было аномальное значение, теперь на первый взгляд пустота. Проверим, так ли это, с помощью метода isna()

```python
df['runtime_clean'].isna().sum()
```

Количество пустых полей - 0... Посмотрим поближе, что же на самом деле записано в данном поле с помощью индексации датасет['столбец'][индекс]

```python
dataset['runtime_clear'][4997]
```

Записано нулевое строковое значение. Убедитесь в этом, проверив тип данных с помощью функции type()

```python
dataset['runtime_clear'].info()
```

Действительно, это строковое значение. Отфильтруем датасет, исключив все такие значения с помощью оператора !=

```python
dataset = dataset[dataset['runtime_clear']!='']
```

Оценим размеры датасета с помощью атрибута shape

```python
dataset.shape
```

А также проверим тип данных с помощью метода info()

```python
dataset.info()
```

Приведем признак runtime_clear в формат int с помощью метода astype()

```python
dataset = dataset[dataset["runtime_clear"].notna()]
```

Проверим тип данных с помощью метода info()

```python
dataset.info()
```

### *5.2. gross_earn*

По аналогии приведем признак gross_earn к числовому типу данных

Выведем Series ['gross_earn'] для детального просмотра данных

```python
dataset['gross_earn']
```

Напишем функцию, которая исключит нецифровые символы с помощью регулярных выражений. Подумаем, нужно ли оставлять при этом в них точку.

```python
def clear_gross_earn(gross_earn):
    return re.sub(r'[^\d\.]', '', gross_earn)
```

Оценим количество пропущенных значений по признаку с помощью метода isna() в связке с методом sum()

```python
dataset['gross_earn'].isna().sum()
```

Необходимо добавить в функцию обработку пустых значений. Для этого импортируем библиотеку numpy с коротким наименованием np

```python
import numpy as np
```

Добавим в функцию проверку на пустые значения с помощью метода pandas pd.isna(). В случае пропуска она должна вернуть пустое значение np.nan; в ином случае - сохранить алгоритм, написанный ранее (переписывать функцию ниже не будем; просто изучим, как она работает)

```python
def clear_gross_earn(gross_earn):
  if pd.isna(gross_earn):
    return np.nan
  else:
    return re.sub('[^\d\.]', '', gross_earn)
```

Примените функцию к Series ['gross_earn'] с помощью метода apply()

```python
dataset['gross_earn'].apply(clear_gross_earn)
```

Создадим столбец gross_earn в исходном датасете, применив функцию к признаку gross_earn

```python
dataset['gross_earn_clear'] = dataset['gross_earn'].apply(clear_gross_earn)
```

Выведем Series ['gross_earn_clear'] для оценки результата

```python
dataset['gross_earn_clear']
```

Приведем признак gross_earn_clear в формат float с помощью метода astype()

```python
dataset['gross_earn_clear'] = dataset['gross_earn_clear'].astype(float)
```

Код был выполнен. Оценим результат, детально рассмотрев Series ['gross_earn_clear']

```python
dataset['gross_earn_clear']
```

Выведем информацию о датасете с помощью метода info()

```python
dataset.info()
```

Все получилось: признаки runtime и gross_earn в числовом формате!
